```python
def mon_to_digit(month):
    if month=='Jan':
        d='01'
    elif month=='Feb':
        d='02'
    elif month=='Mar':
        d='03'
    elif month=='Apr':
        d='04'
    elif month=='May':
        d='05'
    elif month=='Jun':
        d='06'
    elif month=='Jul':
        d='07'
    elif month=='Aug':
        d='08'
    elif month=='Sep':
        d='09'
    elif month=='Oct':
        d='10'
    elif month=='Nov':
        d='11'
    elif month=='Dec':
        d='12'
    return d
```
