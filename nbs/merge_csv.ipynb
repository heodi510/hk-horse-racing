```python
# merge data from each season

path_data= Path.cwd().parent/'data/raw'
years = ['16','17','18','19','20','21']
seasons = [ years[i]+years[i+1] for i in range(len(years)) if i <len(years)-1]
df = [pd.read_csv(path_data/f'merge_s{season}.csv', low_memory=False) for season in seasons]
```

```python
df_concat = pd.concat(df,axis=0)
```

```python
df_concat=df_concat.reset_index(drop=True)
```
